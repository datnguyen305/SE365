```python
import sys
import os
# Thêm thư mục gốc của repo vào sys.path để import các module tự viết
repo_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
if repo_path not in sys.path:
    sys.path.append(repo_path)
print('Đã thêm repo vào sys.path:', repo_path)
```
